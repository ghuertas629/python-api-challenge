```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import scipy.stats as st
import json

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
print(len(cities))
print(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
units = 'Imperial'

url = "http://api.openweathermap.org/data/2.5/weather?appid="

city_url = url + weather_api_key + '&units=' + units + '&q='

print('Beginning Data Retrieval')
print('------------------------')

city_name = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []
count = 0

for city in cities:
    url_response = requests.get(city_url + city)
    url_json = url_response.json()
    try:
        city_name.append(url_json['name'])
        cloudiness.append(url_json['clouds']['all'])
        country.append(url_json['sys']['country'])
        date.append(url_json['dt'])
        humidity.append(url_json['main']['humidity'])
        lat.append(url_json['coord']['lat'])
        lng.append(url_json['coord']['lon'])
        max_temp.append(url_json['main']['temp_max'])
        wind_speed.append(url_json['wind']['speed'])
        count = count + 1
        print(f'Processing Record {count} of Set 1 | {city}')
        time.sleep(1)
    except:
        print('City not found. Skipping...')
        
print('------------------------')
print('Data Retrieval Complete')
print('------------------------')
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
weather_data_df = pd.DataFrame({'City': city_name,
                                'Cloudiness': cloudiness,
                                'Country': country,
                                'Date': date,
                                'Humidity': humidity,
                                'Lat': lat,
                                'Lng': lng,
                                'Max Temp': max_temp,
                                'Wind Speed': wind_speed                               
                               })

weather_data_df.to_csv('Output_Files/Weather_Data.csv', index=False, header=True)
```

## Linear Regression

```python
# OPTIONAL: Create a function to create Linear Regression plots
def lin_reg_plot(x_axis, y_axis, y_label, grid_coordinates, title):
    lin_slope, lin_int, lin_r, lin_p, lin_std_err = st.linregress(x_axis, y_axis)
    lin_eq = "y = " + str(round(lin_slope,2)) + "x + " + str(round(lin_int,2))
    
    plt.scatter(x_axis, y_axis, marker="o", facecolors="red", edgecolors="black")
    lin_reg = lin_slope * x_axis + lin_int
    plt.plot(x_axis, lin_reg, color='black')
    plt.annotate(lin_eq, grid_coordinates, fontsize=14, color='black')
    plt.xlabel('Latitude')
    plt.ylabel(y_label)
    plt.title(title)
    print(f"The r-squared is: {round(lin_r,2)}")
    file_name = f'Output_Files/{title}.png'
    plt.savefig(file_name)
    plt.show()
```
